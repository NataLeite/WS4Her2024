### A Torre do Feiticeiro üßôüèª‚Äç‚ôÄÔ∏èü™Ñ


O poderoso feiticeiro Zargof construiu uma torre m√°gica com v√°rias portas trancadas. Cada porta possui uma chave m√°gica √∫nica. Por√©m, as portas e as chaves t√™m uma peculiaridade: a torre s√≥ pode ser atravessada se voc√™ **abrir as portas na ordem inversa da sequ√™ncia em que as encontrou.**<br>

Quando voc√™ entra na torre, voc√™ come√ßa com um invent√°rio vazio.
Ao encontrar uma porta, voc√™ pega sua respectiva chave e guarda.
Para sair da torre, voc√™ precisa abrir todas as portas na ordem inversa usando as chaves que guardou.<br>

Voc√™ recebe a sequ√™ncia de todas as portas na ordem em que foram encontradas (desde o come√ßo at√© sua suposta sa√≠da). Escreva um programa que determine se foi poss√≠vel sair da torre seguindo a regra.
<br><br>
**Entrada** <br>
Uma string composta apenas por caracteres alfab√©ticos mai√∫sculos, onde cada letra representa uma porta. Por exemplo, `ABBA` significa que voc√™ encontrou as portas `A`, `B`, `B` e `A`, nesta ordem. <br><br>

**Sa√≠da** <br>
Imprima `YES` se for poss√≠vel sair da torre, seguindo a regra, ou `NO` caso contr√°rio.
<br><br>
**Exemplos** <br>

Entrada 1:
```
ABBA
```
Sa√≠da 1:
```
YES
```
<br>

Entrada 2:
```
ABAB
```
Sa√≠da 2:
```
NO
```
<br>


````python
# defina sua fun√ß√£o aqui
def can_exit_tower(sequence:str) -> str:
    pilha = Pilha()
    for porta in sequence:
        if ~pilha.esta_vazia() and pilha.peek() == porta:  # Porta corresponde √† √∫ltima empilhada
            pilha.pop()
        else:  # Adiciona nova porta √† pilha
            pilha.push(porta)
    return "YES" if pilha.esta_vazia() else "NO"


class Pilha:
    def __init__(self):
        self.itens = []

    def push(self, item):  # adiciona um item ao topo da pilha.
        self.itens.append(item)

    def pop(self):   # remove e retorna o item do topo da pilha.
        if not self.esta_vazia():
            return self.itens.pop()
        else:
            return "A pilha est√° vazia!"

    def peek(self):   # retorna o item do topo da pilha sem remov√™-lo.
        if not self.esta_vazia():
            return self.itens[-1]
        else:
            return "A pilha est√° vazia!"

    def esta_vazia(self):  # verifica se a pilha est√° vazia para o usu√°rio.
        return len(self.itens) == 0

    def tamanho(self):   # retorna o tamanho da pilha
        return len(self.itens)
````

### A Escadaria da Hero√≠na ü¶∏‚Äç‚ôÄÔ∏è

A hero√≠na Aria est√° subindo uma escadaria para alcan√ßar o templo sagrado. A escadaria tem `ùëÅ` degraus, e Aria pode subir de 1 ou 2 degraus por vez. ‚ÜóÔ∏è

Sua tarefa √© determinar **de quantas formas diferentes Aria pode alcan√ßar o topo** da escadaria (<u>o degrau ùëÅ</u>).
<br><br>
**Regras**
1. Aria come√ßa no ch√£o (degrau `0`).
2. Ela s√≥ pode avan√ßar `1` ou `2` degraus por vez.
3. Calcule o n√∫mero total de maneiras que ela pode alcan√ßar o degrau `ùëÅ`.

<br>**Entrada**<br>
Um n√∫mero inteiro `ùëÅ` (1 ‚â§ ùëÅ ‚â§ 30) representando o n√∫mero total de degraus.
<br><br>
**Sa√≠da**<br>
Um √∫nico n√∫mero inteiro representando o n√∫mero de maneiras de alcan√ßar o topo.

<br>

**Exemplos**

<br>

Entrada 1:


```
2
```
Sa√≠da 1:


```
2
```
<br><br>
Entrada 2:

````python
def formas_de_subir(N:int)-> int:
    if N == 0:  # Caso base: n√£o h√° degraus
        return 1
    if N == 1:  # Caso base: apenas um degrau
        return 1
    # Passo recursivo: soma das formas de subir de N-1 e N-2
    return formas_de_subir(N - 1) + formas_de_subir(N - 2)
````

#### üí° Preciso de uma dica!

Para resolver o problema de forma recursiva, pense no seguinte:

1. Identifique os casos base:

Se ùëÅ = 0: Aria j√° est√° no topo, h√° exatamente 1 maneira de alcan√ßar o degrau 0.
Se ùëÅ = 1: H√° apenas 1 maneira de alcan√ßar o primeiro degrau (um √∫nico passo de 1 degrau).

2. Divida o problema em subproblemas menores:

Para alcan√ßar o degrau ùëÅ, Aria pode:
 1. Subir 1 degrau e partir da√≠ pra frente com ùëÅ‚àí1 degraus.
 2. Subir 2 degraus e da√≠ pra frente subir ùëÅ‚àí2 degraus.

Assim, a solu√ß√£o para ùëÅ √© a soma das solu√ß√µes para ùëÅ‚àí1 e ùëÅ‚àí2:


```
n_jeitos(ùëÅ) = n_jeitos(ùëÅ‚àí1) + n_jeitos(ùëÅ‚àí2)

```


Implemente a recurs√£o:

Escreva uma fun√ß√£o recursiva que calcula o n√∫mero de maneiras para ùëÅ com base nos **casos base** e na **rela√ß√£o** acima.

#### Desafio Extra üî•

Embora a recurs√£o pura funcione para ùëÅ pequeno, ela pode se tornar ineficiente para ùëÅ maiores devido √† **repeti√ß√£o de c√°lculos**. Para evitar isso, podemos usar uma t√©cnica chamada **memoiza√ß√£o**. Que tal pesquisar um pouco sobre ela e tentar reimplementar sua fun√ß√£o, comparando os tempos de execu√ß√£o das duas vers√µes para ùëÅ muitos grandes?

````python
# reimplemente sua fun√ß√£o aqui
def formas_de_subir_memoizada(N:int, memo=None)-> int:
    if memo is None:
        memo = {}  # Inicializa o dicion√°rio de memoiza√ß√£o

    if N == 0:  # Caso base: n√£o h√° degraus
        return 1
    if N == 1:  # Caso base: apenas um degrau
        return 1

    if N in memo:  # Retorna o valor armazenado, se j√° calculado
        return memo[N]

    # Passo recursivo com memoiza√ß√£o
    memo[N] = formas_de_subir_memoizada(N - 1, memo) + formas_de_subir_memoizada(N - 2, memo)
    return memo[N]
````

### O Labirinto do Mago ‚ÜóÔ∏è‚ÜòÔ∏è




O mago Althar ficou preso em um labirinto m√°gico com salas conectadas por portais unidirecionais. Cada sala cont√©m uma energia m√°gica que o mago pode absorver, mas ele s√≥ pode passar por cada sala uma √∫nica vez.

O objetivo √© ajudar o mago a encontrar o caminho que **maximiza a energia total coletada**, come√ßando na sala 0.
<br><br>

**Regras**
1. Cada sala √© representada por um n√∫mero (de `0` a `N-1`).
2. Uma lista de inteiros `energia`, onde `energia[i]` representa a energia dispon√≠vel na sala `i`.
3. Uma matriz `portais`, onde `portais[i]` √© uma lista das conex√µes da sala `i`, isto √©, cada elemento da lista `portais[i]` √© o √≠ndice de uma sala acess√≠vel a partir de `i`.
4. O mago pode parar em qualquer sala ou continuar explorando at√© que n√£o possa mais se mover.

<br>

**Entrada**

2. Uma lista `energia` de tamanho `ùëÅ`.
2. Uma lista de listas `portais`, onde `portais[i]` cont√©m os √≠ndices das salas acess√≠veis a partir da sala `ùëñ`.
<br>
**Sa√≠da**<br>

<br><br>
**Exemplos**

````python
def calcular_energia_maxima_recursiva(sala, energia, portais):
    """
    Fun√ß√£o recursiva para calcular a energia m√°xima colet√°vel.
    """
    # Caso base: se n√£o houver portais saindo da sala, retorna apenas a energia da sala atual
    if not portais[sala]:
        return energia[sala]

    # Calcula a energia m√°xima colet√°vel a partir da sala atual
    max_energia = 0
    for prox_sala in portais[sala]:
        max_energia = max(max_energia, calcular_energia_maxima_recursiva(prox_sala, energia, portais))

    # Soma a energia da sala atual com o melhor caminho encontrado
    return energia[sala] + max_energia

def calcular_energia_maxima(N, energia, portais):
    max_energia = 0
    max_energia = max(max_energia, calcular_energia_maxima_recursiva(0, energia, portais))
    return max_energia
````

````python
# resolu√ß√£o usando memoriza√ß√£o
def max_energia(sala, energia, portais, visitado, memo):
    # Se j√° calculamos a melhor energia para esta sala, retornamos
    if sala in memo:
        return memo[sala]

    # Marca a sala como visitada
    visitado[sala] = True
    max_energy = energia[sala]

    # Explora todos os portais acess√≠veis a partir desta sala
    for prox_sala in portais[sala]:
        if not visitado[prox_sala]:
            max_energy = max(max_energy, energia[sala] + max_energia(prox_sala, energia, portais, visitado, memo))

    # Marca como n√£o visitado (backtracking)
    visitado[sala] = False

    # Armazena no memo e retorna
    memo[sala] = max_energy
    return max_energy

def resolver_labirinto(N, energia, portais):
    visitado = [False] * N
    memo = {}
    return max_energia(0, energia, portais, visitado, memo)
````

### Encontre o Menor Lan√ßa-Foguetes üéÜ

O reino est√° se preparando para um grande festival de fogos de artif√≠cio, mas os foguetes s√£o extremamente potentes e s√≥ podem ser lan√ßados de uma altura m√≠nima especificada para garantir a seguran√ßa.

Dada uma lista de alturas dispon√≠veis para construir lan√ßa-foguetes, encontre a menor altura que seja maior ou igual √† altura m√≠nima necess√°ria.

<br>

**Entrada**
<br>
1. Uma lista `alturas` de ùëÅ inteiros positivos, representando as alturas **ordenadas em ordem decrescente**.
2. Um inteiro `ùêª`, a altura m√≠nima necess√°ria.
<br>

**Sa√≠da**

Imprima a altura do menor lan√ßa-foguetes que atende √† altura m√≠nima ùêª.
Se nenhuma altura atender √† condi√ß√£o, imprima `-1`.

<br>

**Exemplos**

Entrada 1:

```
[50, 40, 30, 20, 10]
25
```
Sa√≠da 1:
```
30
```
<br>

Entrada 2:
```
[20, 15, 10, 5]
21

```
Sa√≠da 2:
```
-1
```

````python
def menor_altura(alturas, H):
    inicio = 0
    fim = len(alturas) - 1
    resposta = -1

    while inicio <= fim:
        meio = (inicio + fim) // 2

        if alturas[meio] >= H:
            resposta = alturas[meio]  # Candidata v√°lida
            inicio = meio + 1  # Continua buscando uma menor altura v√°lida
        else:
            fim = meio - 1  # Altura √© muito baixa, busca √† esquerda

    return resposta
````

````python
#@title Rode para testar sua fun√ß√£o! (ctrl+enter)

from time import time
import pandas as pd

test_cases = pd.read_csv('https://raw.githubusercontent.com/maris-silva/testcases-w4h/refs/heads/main/test_cases_lanca_foguetes.csv')
test_cases['Alturas'] = test_cases['Alturas'].apply(eval)

passed = True
start_time = time()
for i, data in test_cases.iterrows():
      data = data.tolist()
      alturas, H, expected = data[0], data[1], data[2]
      result = menor_altura(alturas, H)
      if result != expected:
          print(f"  Input: alturas = {alturas}, H = {H}, Expected: {expected}, Got: {result}")
          passed = False
end_time = time()
execution_time = end_time - start_time
if passed:
  print('Parab√©ns!! Seu c√≥digo passou por todos os testes, v√° buscar seu bal√£o!')
print(f"Tempo de execu√ß√£o de todos os test cases: {round(execution_time, 3)} segundos")
````

#### üí° Preciso de uma dica!

Como nossa lista est√° **ordenada**, podemos usar um algoritmo de busca - como a busca bin√°ria - para resolv√™-lo de maneira eficiente.

Um ponto a se tomar cuidado aqui √© que a lista est√° ordenada de maneira **decrescente**! Ent√£o voc√™ pode escolher reverter a lista e usar o algoritmo tradicional passado em aula, ou adapt√°-lo para que funcione para lista decrescente (a segunda op√ß√£o √© mais interessante, vamos ver se a ideia do algoritmo ficou clara! üôÇ).

**Dicas pra implementa√ß√£o:**
1. **Defina os limites da busca:** Use dois ponteiros: in√≠cio (in√≠cio da lista) e fim (fim da lista).
2. **Passo na busca bin√°ria:** No meio da lista, verifique se a altura atual atende √† condi√ß√£o (>= H). Se atender, registre essa altura como candidata e continue √† direita para buscar uma menor altura que tamb√©m atenda. Se n√£o atender, mova para a esquerda para buscar alturas maiores.
3. **Condi√ß√£o de parada:** Quando inicio > fim, a busca termina.
4. **Caso nenhuma altura atenda:** Retorne -1 se nenhum valor for encontrado.
